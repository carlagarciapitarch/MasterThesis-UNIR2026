```python
# PyPept modules
from pyPept.sequence import Sequence
from pyPept.sequence import correct_pdb_atoms
from pyPept.molecule import Molecule
from pyPept.converter import Converter
from pyPept.conformer import Conformer
from pyPept.conformer import SecStructPredictor

# RDKit modules
from rdkit import Chem
from rdkit.Chem import Draw
```

```python
def sec_to_smiles_biln(biln):
    # Crear secuencia
    seq = Sequence(biln)

    # Corregir átomos
    seq = correct_pdb_atoms(seq)

    # Crear molécula
    mol = Molecule(seq)
    romol = mol.get_molecule(fmt='ROMol')

    # Obtener SMILES
    smiles = Chem.MolToSmiles(romol)

    return smiles


resultado = sec_to_smiles_biln("A-A-C-T-A-C-T-M-K")
print(resultado)
```

```python
def sec_to_smiles_helm(helm):
    b = Converter(helm=helm)
    biln = b.get_biln()
    seq = Sequence(biln)
    # Correct atom names in the sequence object
    seq = correct_pdb_atoms(seq)

    # Loop wit the included monomers
    mm_list = seq.s_monomers
    for i, monomer in enumerate(mm_list):
        mon = monomer['m_romol']

    mol = Molecule(seq)
    romol = mol.get_molecule(fmt='ROMol')
    smiles = Chem.MolToSmiles(romol)

    return smiles


resultado = sec_to_smiles_helm("PEPTIDE1{[ac].D.T.H.F.E.I.A.[am]}$$$$V2.0")
print(resultado)
```

```python
def peptido_con_guion(peptido):
    return "-".join(peptido)

# Ejemplo de uso
resultado = peptido_con_guion("hola")
print(resultado)  # h-o-l-a
```
